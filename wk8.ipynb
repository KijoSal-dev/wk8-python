```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace with your actual data loading)
# Example: df = pd.read_csv('covid_data.csv')
data = {
    'country': ['United States', 'United States', 'India', 'India', 'Brazil', 'Brazil'],
    'date': ['2020-01-01', '2020-01-02', '2020-01-01', '2020-01-02', '2020-01-01', '2020-01-02'],
    'total_cases': [100, 200, 50, 150, 80, 180],
    'total_deaths': [10, 20, 5, 15, 8, 18],
    'new_cases': [100, 100, 50, 100, 80, 100]
}
df = pd.DataFrame(data)

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Check for required columns
required_columns = ['country', 'date', 'total_cases', 'total_deaths', 'new_cases']
if not all(col in df.columns for col in required_columns):
    missing = [col for col in required_columns if col not in df.columns]
    raise ValueError(f"Missing columns: {missing}")

# Handle missing values
df = df.dropna(subset=['country', 'date', 'total_cases', 'total_deaths'])  # Drop rows with missing critical values
print(f"Missing values in 'country': {df['country'].isna().sum()}")
print(f"Missing values in 'date': {df['date'].isna().sum()}")

# Define selected countries
selected_countries = ['United States', 'India', 'Brazil']  # Adjust based on df['country'].unique()
print("Available countries:", df['country'].unique())
```

```python
def plot_total_cases(df, countries):
    plt.figure(figsize=(12, 6))
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        plt.plot(country_data['date'], country_data['total_cases'], label=country)
    plt.xlabel('Date')
    plt.ylabel('Total Cases')
    plt.title(f'Total COVID-19 Cases Over Time ({", ".join(countries)})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Call the function
plot_total_cases(df, selected_countries)
```

```python
def plot_total_deaths(df, countries):
    plt.figure(figsize=(12, 6))
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        plt.plot(country_data['date'], country_data['total_deaths'], label=country)
    plt.xlabel('Date')
    plt.ylabel('Total Deaths')
    plt.title(f'Total COVID-19 Deaths Over Time ({", ".join(countries)})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Call the function
plot_total_deaths(df, selected_countries)
```

```python
def plot_new_cases(df, countries):
    plt.figure(figsize=(12, 6))
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        plt.plot(country_data['date'], country_data['new_cases'], label=country)
    plt.xlabel('Date')
    plt.ylabel('Daily New Cases')
    plt.title(f'Daily New COVID-19 Cases Over Time ({", ".join(countries)})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Call the function
plot_new_cases(df, selected_countries)
```

```python
def calculate_death_rate(df, countries):
    death_rates = {}
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        # Use the latest available data
        latest_data = country_data.iloc[-1]
        if latest_data['total_cases'] > 0:  # Avoid division by zero
            death_rate = latest_data['total_deaths'] / latest_data['total_cases']
            death_rates[country] = death_rate
        else:
            death_rates[country] = 0
    return death_rates

# Calculate and print death rates
death_rates = calculate_death_rate(df, selected_countries)
for country, rate in death_rates.items():
    print(f"Death rate for {country}: {rate:.4f} ({rate*100:.2f}%)")
```

```python
def plot_top_countries_by_cases(df, n=5):
    # Get the latest total cases for each country
    latest_data = df.groupby('country').apply(lambda x: x.sort_values('date').iloc[-1])
    top_countries = latest_data.sort_values('total_cases', ascending=False).head(n)
    
    plt.figure(figsize=(10, 6))
    plt.bar(top_countries.index, top_countries['total_cases'])
    plt.xlabel('Country')
    plt.ylabel('Total Cases')
    plt.title(f'Top {n} Countries by Total COVID-19 Cases')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Call the function
plot_top_countries_by_cases(df, n=5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace with your actual data loading)
# Example: df = pd.read_csv('vaccination_data.csv')
data = {
    'country': ['United States', 'United States', 'India', 'India', 'Brazil', 'Brazil'],
    'date': ['2021-01-01', '2021-01-02', '2021-01-01', '2021-01-02', '2021-01-01', '2021-01-02'],
    'total_vaccinations': [1000, 2000, 500, 1500, 800, 1800],
    'people_fully_vaccinated': [500, 1000, 200, 600, 300, 900],
    'population': [331000000, 331000000, 1380000000, 1380000000, 212000000, 212000000]
}
df = pd.DataFrame(data)

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Check for required columns
required_columns = ['country', 'date', 'total_vaccinations', 'people_fully_vaccinated', 'population']
if not all(col in df.columns for col in required_columns):
    missing = [col for col in required_columns if col not in df.columns]
    raise ValueError(f"Missing columns: {missing}")

# Handle missing values
df = df.dropna(subset=['country', 'date', 'total_vaccinations', 'people_fully_vaccinated'])
print(f"Missing values in 'country': {df['country'].isna().sum()}")
print(f"Missing values in 'date': {df['date'].isna().sum()}")
print(f"Missing values in 'total_vaccinations': {df['total_vaccinations'].isna().sum()}")
print(f"Missing values in 'people_fully_vaccinated': {df['people_fully_vaccinated'].isna().sum()}")

# Define selected countries
selected_countries = ['United States', 'India', 'Brazil']  # Adjust based on df['country'].unique()
print("Available countries:", df['country'].unique())
```

```python
def plot_cumulative_vaccinations(df, countries):
    plt.figure(figsize=(12, 6))
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)
    plt.xlabel('Date')
    plt.ylabel('Cumulative Vaccinations')
    plt.title(f'Cumulative COVID-19 Vaccinations Over Time ({", ".join(countries)})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Call the function
plot_cumulative_vaccinations(df, selected_countries)
```

```python
def plot_percent_vaccinated(df, countries):
    plt.figure(figsize=(12, 6))
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        percent_vaccinated = (country_data['people_fully_vaccinated'] / country_data['population']) * 100
        plt.plot(country_data['date'], percent_vaccinated, label=country)
    plt.xlabel('Date')
    plt.ylabel('Percentage Fully Vaccinated (%)')
    plt.title(f'Percentage of Population Fully Vaccinated Over Time ({", ".join(countries)})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Call the function
plot_percent_vaccinated(df, selected_countries)
```

```python
def plot_percent_vaccinated_bar(df, countries):
    latest_percent = []
    valid_countries = []
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        latest_data = country_data.iloc[-1]
        percent_vaccinated = (latest_data['people_fully_vaccinated'] / latest_data['population']) * 100
        latest_percent.append(percent_vaccinated)
        valid_countries.append(country)
    
    plt.figure(figsize=(10, 6))
    plt.bar(valid_countries, latest_percent)
    plt.xlabel('Country')
    plt.ylabel('Percentage Fully Vaccinated (%)')
    plt.title('Latest Percentage of Population Fully Vaccinated')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Call the function
plot_percent_vaccinated_bar(df, selected_countries)
```

```python
def plot_vaccination_pie_charts(df, countries):
    for country in countries:
        if country not in df['country'].values:
            print(f"Warning: No data for {country}")
            continue
        country_data = df[df['country'] == country].sort_values('date')
        if country_data.empty:
            print(f"No data available for {country}")
            continue
        latest_data = country_data.iloc[-1]
        vaccinated = latest_data['people_fully_vaccinated']
        unvaccinated = latest_data['population'] - vaccinated
        
        plt.figure(figsize=(6, 6))
        plt.pie([vaccinated, unvaccinated], labels=['Fully Vaccinated', 'Unvaccinated'], autopct='%1.1f%%')
        plt.title(f'Vaccination Status in {country} (Latest Data)')
        plt.show()

# Call the function
plot_vaccination_pie_charts(df, selected_countries)
```
